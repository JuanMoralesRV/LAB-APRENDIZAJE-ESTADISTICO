# STATSMODEL.API

```python
#Cargar df
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("HousingData.csv")
df.describe()
```

## Limpieza Datos

```python
#Quitar valores nulos
df_2 = df.dropna()

df_2.describe()
```

```python
#Quitar valores duplicados
df_3 = df_2.drop_duplicates()

df_3.describe()
```

## Modelo

```python
import statsmodels.api as sm
import numpy as np

CRIM = df_3["CRIM"].values.reshape([-1, 1])
ZN = df_3["ZN"].values.reshape([-1, 1])
INDUS = df_3["INDUS"].values.reshape([-1, 1])
CHAS = df_3["CHAS"].values.reshape([-1, 1])
NOX = df_3["NOX"].values.reshape([-1, 1])
RM = df_3["RM"].values.reshape([-1, 1])
AGE = df_3["AGE"].values.reshape([-1, 1])
DIS = df_3["DIS"].values.reshape([-1, 1])
RAD = df_3["RAD"].values.reshape([-1, 1])
TAX = df_3["TAX"].values.reshape([-1, 1])
PTRATIO = df_3["PTRATIO"].values.reshape([-1, 1])
B = df_3["B"].values.reshape([-1, 1])
LSTAT = df_3["LSTAT"].values.reshape([-1, 1])

n = len(df_3)

unos = np.ones([n,1])

x = np.hstack([
    unos,
    CRIM,
    ZN,
    INDUS,
    CHAS,
    NOX,
    RM,
    AGE,
    DIS,
    RAD,
    TAX,
    PTRATIO,
    B,
    LSTAT
])

y = df_3["MEDV"].values.reshape([-1,1])

ols = sm.OLS(y, x) #fit es al reves, es decir, primero Y y luego X

results = ols.fit()

results.summary()
```

## Escalamiento / estandarizar

```python
for column in df_3.columns:
    col_min = np.min(df_3[column])
    col_max = np.max(df_3[column])
    print(f"{column}: min = {col_min}, max = {col_max}")
```

```python
df_4 = df_3

for column in df_4.columns:
    if column != "MEDV":  # No escalar la variable objetivo
        col_min = df_4[column].min()
        col_max = df_4[column].max()
        df_4[column] = (df_4[column] - col_min) / (col_max - col_min)
```
