# Factores Cualitativos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dummies

Se usa [0,1] para indicar ausencia de o no de algo

Se usa [-1,1] para indicar quitar o poner algo 
```

# Datos

```python
df = pd.read_excel("factores cualitativos.xlsx")

calificaciones = [9,7,10,8,8,6,10,7,8,6,9.5,7,9,7,10,8]

df["Calificación"] = calificaciones
```

# EDA

```python
df.head(16)
```

# Modelo

```python
import statsmodels.api as sm

X = df.drop(columns=["Calificación"])
y = df["Calificación"]

# Agregar columna de unos para el intercepto
X = sm.add_constant(X)

# Crear el modelo OLS
ols = sm.OLS(y, X)

# Ajustar el modelo
results = ols.fit()

# Resumen
results.summary()
```
