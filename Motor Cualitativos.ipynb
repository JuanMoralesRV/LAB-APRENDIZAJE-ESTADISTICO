# Train / Test

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns = ["model", "mpg"])
y = df['mpg']  
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

# Escalamiento de variables

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X_train)

X_scaled_train = scaler.transform(X_train)

X_scaled_test = scaler.transform(X_test)
```

# Modelo

```python

import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm

# Agregar constante para el término independiente
X_scaled_train_const = sm.add_constant(X_scaled_train)
X_scaled_test_const = sm.add_constant(X_scaled_test)
```

```python
# Modelo train
modelo_train = sm.OLS(y_train, X_scaled_train_const).fit()

modelo_train.summary()
```

```python
# 7. Evaluación del modelo entrenado con TRAIN en TEST

y_pred = modelo_train.predict(X_scaled_test_const)

r2 = r2_score(y_test, y_pred)

print(f"R²: {r2:.4f}")
```
