###  Realiza un train-test-split donde se use el 40% de los datos para entrenar. Calcula el R2 de entrenamiento y de prueba.

```python
#Definir "x", "y"/ Separar en train y test

from sklearn.model_selection import train_test_split

X = df.drop(columns = ["model", "mpg"])
y = df['mpg']  

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.4, random_state=137)
```

```python
# Escalamiento de variables x

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X_train)

X_scaled_train = scaler.transform(X_train)

X_scaled_test = scaler.transform(X_test)
```

```python
# Regresión

from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm

# Agregar constante para el término independiente
X_scaled_train_const = sm.add_constant(X_scaled_train)
X_scaled_test_const = sm.add_constant(X_scaled_test)

# Modelo train
modelo_train = sm.OLS(y_train, X_scaled_train_const).fit()

print(modelo_train.summary())

# Evaluación del modelo entrenado con TRAIN en TEST

y_pred_test = modelo_train.predict(X_scaled_test_const)

r2 = r2_score(y_test, y_pred_test)


print("")
print("")
print("")
print("")
print(f"R² de test: {r2:.4f}")
```

### Añade regularización L2 con un hiperparámetro lambda decidido por ti. Cambia este valor y compara con varios distintos los R2 de entrenamiento y de prueba.

```python
from sklearn.linear_model import Ridge

# Entrenar el modelo Ridge 
modelo_ridge = Ridge(alpha=16)
modelo_ridge.fit(X_scaled_train, y_train)

# Predicciones
y_train_pred_ridge = modelo_ridge.predict(X_scaled_train)
y_test_pred_ridge = modelo_ridge.predict(X_scaled_test)

# R2
r2_train_ridge = r2_score(y_train, y_train_pred_ridge)
r2_test_ridge = r2_score(y_test, y_test_pred_ridge)

# Resultados Ridge
print("Coeficientes Ridge:", modelo_ridge.coef_)
print("Intercepto Ridge:", modelo_ridge.intercept_)
print(f"R² en entrenamiento: {r2_train_ridge:.4f}")
print(f"R² en prueba:        {r2_test_ridge:.4f}")
```

### Modelo dummies sin train / test

```python
#EDA

import pandas as pd

df3 = pd.read_excel("Motor Trend Car Road Tests.xlsx")

# print(df3.head())
# print(df3.describe())
# print(df3.info())



#Columnas dummies

df3 = pd.get_dummies(df3, columns=["cyl", "gear", "carb"], drop_first=False)

#Definir "x", "y"

from sklearn.model_selection import train_test_split

X = df3.drop(columns = ["model", "mpg"])
y = df3['mpg'] 

# Escalamiento de variables x

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)

X_scaled = scaler.transform(X)

# Regresión

from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm

# Agregar constante para el término independiente
X_scaled_const = sm.add_constant(X_scaled)

# Modelo 21
modelo_21 = sm.OLS(y, X_scaled_const).fit()

print(modelo_21.summary())
```

### Modelo dummies con train / test

```python
#columnas dummies
df3 = pd.read_excel("Motor Trend Car Road Tests.xlsx")
df3 = pd.get_dummies(df3, columns=["cyl", "gear", "carb"], drop_first=False)

#definir "x" "y"
X = df3.drop(columns=["model", "mpg"])
y = df3["mpg"]

#separar en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.4, random_state=137)

#scalar variables
scaler = StandardScaler().fit(X_train)
X_scaled_train = scaler.transform(X_train)
X_scaled_test = scaler.transform(X_test)


#convertir array a df
X_scaled_train_df = pd.DataFrame(X_scaled_train, index=X_train.index, columns=X_train.columns)
X_scaled_test_df = pd.DataFrame(X_scaled_test, index=X_test.index, columns=X_test.columns)

#añadir constante para el modelo OLS
X_scaled_train_const = sm.add_constant(X_scaled_train_df)
X_scaled_test_const = sm.add_constant(X_scaled_test_df)

# Alinear columnas test a train 
X_scaled_test_const = X_scaled_test_const.reindex(columns=X_scaled_train_const.columns, fill_value=0)

#modelo
modelo_train = sm.OLS(y_train, X_scaled_train_const).fit()

print(modelo_train.summary())

#modelo train evaluado en test
y_pred_test = modelo_train.predict(X_scaled_test_const)

r2 = r2_score(y_test, y_pred_test)
print(f"\nR² de test: {r2:.4f}")
```
